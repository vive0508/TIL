### 1.2.2 그래프 그리기 기초 예제

#### 1.2.2.1 삼각함수 그리기
- np.arange(a, b, s): a부터 b까지 s의 간격
- np.sin(value)

```python
import numpy as np 

x = np.arange(0, 10, 0.01)
y = np.sin(x)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

```python
def drawGraph():
    plt.figure(figsize=(10, 5))
    plt.plot(x, np.sin(x), label="sin")
    plt.plot(x, np.cos(x), label="cos")
    plt.grid(True) # 격자무늬 
    plt.legend(loc='upper right') # 범례(위치)
    plt.title("Sample") # 제목
    plt.xlabel("time") # x축 제목
    plt.ylabel("Amplitude") # y축 제목 
    plt.show()
    
drawGraph()
```

#### 1.2.2.2 그래프 커스텀

```python
x = np.arange(0, 5, 0.5)
x
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, x, "r--") # red ---- 
plt.plot(x, x ** 2, "bs") # blue square
plt.plot(x, x ** 3, "g>") # green >
plt.grid(True)
plt.show()
```

```python
# t = [0, 1, 2, 3, 4, 5, 6]
x = list(range(0, 7))
y = [1, 4, 5, 8, 9, 5, 3]
```

```python
def drawGraph():

    plt.figure(figsize=(10, 6))
    plt.plot(
        x,
        y,
        color="red", 
        linestyle="--", #'dashed','-'
        marker="o", 
        markerfacecolor="blue",
        markersize=10, 
    )

    plt.grid(True)
    
    # x축과, y축의 범위 지정
    plt.xlim([-0.5, 6.5]) 
    plt.ylim([0.5, 9.5])
    plt.show() 
    
drawGraph()
```

#### 1.2.2.3 scatter plot

```python
x = np.array(range(0, 10))
y = np.array([9, 8, 7, 9, 8, 3, 2, 4, 3, 4])
```

```python
def drawGraph():

    plt.figure(figsize=(10, 6))
    plt.scatter(x, y)
    plt.grid(True)
    plt.show()
    
drawGraph()
```
